```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import json

with open("all_raw_results_cpu.json", "rt") as f:
  all_raw_results = json.load(f)

all_raw_results = {outer_key: {int(inner_key): inner_value for inner_key, inner_value in inner_dict.items()}
        for outer_key, inner_dict in all_raw_results.items()}

average_df = pd.read_csv("average_df_cpu.csv", index_col='Format')

iterations=1
```

```python
from datasets import load_dataset

ds = load_dataset("HichTala/dota")
```

```python
from tqdm import tqdm
import os
from datasets import load_dataset

# Создание директории images внутри /content
image_dir = "images"  #  Относительный путь!
os.makedirs(image_dir, exist_ok=True)

# Извлечение и сохранение ТОЛЬКО тестовых изображений
image_paths = []
for i, example in enumerate(tqdm(ds['test'], desc="Saving test images")):
    image = example['image']
    image_path = os.path.join(image_dir, f"test_image_{i}.jpg")  # Относительный путь
    image.save(image_path)
    image_paths.append(image_path)

image_paths = image_paths[:150]
```

```python
# 1. Столбчатая диаграмма (Bar Chart) для сравнения среднего времени инференса

plt.figure(figsize=(10, 6))

# Используем average_df напрямую
bars = plt.bar(average_df.index, average_df['Total Inference Time (s)'], color='skyblue', edgecolor='grey')

# Добавляем подписи значений НАД столбцами
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.001,  # Небольшой сдвиг вверх
             f"{yval:.3f}",  # Форматируем до 3 знаков после запятой
             ha='center',  # Центрируем по горизонтали
             va='bottom',   # Выравниваем по нижнему краю
             fontsize=10,
             color='black')

# Добавляем подписи осей и заголовок
plt.xlabel('Format', fontweight='bold')
plt.ylabel('Total Inference Time (s)', fontweight='bold')
plt.title('Comparison of Average Inference Time', fontweight='bold')

# Добавляем горизонтальную сетку
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.ylim(0, average_df['Total Inference Time (s)'].max() * 1.15)

plt.tight_layout()
plt.show()
```

```python
image_number_map = {}  # Словарь для хранения соответствия image_path -> image_number
image_counter = 0

for iteration in range(iterations):
    line_data = []
    for image_path in image_paths:
        if image_path not in image_number_map:
            image_number_map[image_path] = image_counter
            image_counter += 1

        for format, lib_data in all_raw_results.items():
            if iteration in lib_data:
                # Рекурсивная функция для поиска словарей в списках
                def find_dicts(data):
                    if isinstance(data, list):
                        for item in data:
                            yield from find_dicts(item)
                    elif isinstance(data, dict):
                        yield data

                # Ищем результат для данного image_path на данной итерации
                for result in find_dicts(lib_data[iteration]):
                    if result.get('image_path') == image_path:
                        inference_time = result['model_speed']['inference']
                        line_data.append({
                            'format': format,
                            'image_number': image_number_map[image_path],
                            'inference_time': inference_time
                        })
                        break
    line_df = pd.DataFrame(line_data)

    plt.figure(figsize=(12, 6))
    if not line_df.empty:
        # Используем scatterplot вместо lineplot
        sns.scatterplot(data=line_df, x='image_number', y='inference_time', hue='format', s=100, alpha=1)  # s=100 - размер точек

    else:
        print(f"Warning: No data found for iteration {iteration}. Skipping plot.")
        continue

    plt.xlabel('Image Number')
    plt.ylabel('Inference Time (ms)')
    plt.title(f'Inference Time vs. Image Number (Iteration {iteration})')
    plt.legend(title='format')
    plt.grid(axis='y', linestyle='-', alpha=0.7)
    plt.grid(axis='x', linestyle=':', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

image_number_map = {}  # Словарь для хранения соответствия image_path -> image_number
image_counter = 0
formats = all_raw_results.keys() # Получаем список форматов

for format in formats:
    plt.figure(figsize=(12, 6))  # Создаем новую фигуру для каждого формата
    data_for_format = []  # Список для хранения данных для текущего формата

    for iteration in range(iterations):
        for image_path in image_paths:
            if image_path not in image_number_map:
                image_number_map[image_path] = image_counter
                image_counter += 1

            # Проверяем, есть ли данные для текущего формата
            if format in all_raw_results and iteration in all_raw_results[format]:

                # Рекурсивная функция для поиска словарей в списках (если есть вложенность)
                def find_dicts(data):
                    if isinstance(data, list):
                        for item in data:
                            yield from find_dicts(item)
                    elif isinstance(data, dict):
                        yield data

                # Ищем результат для данного image_path на данной итерации
                for result in find_dicts(all_raw_results[format][iteration]):
                    if result.get('image_path') == image_path:
                        inference_time = result['model_speed']['inference']
                        data_for_format.append({
                            'format': format,
                            'image_number': image_number_map[image_path],
                            'inference_time': inference_time
                        })
                        break  # нашли, выходим из цикла по результатам

    # Создаем DataFrame только для текущего формата
    line_df = pd.DataFrame(data_for_format)

    if not line_df.empty:
        # Получаем цвет из стандартной палитры для текущего формата
        color = sns.color_palette()[list(formats).index(format)]

        # Используем scatterplot с фиксированным цветом
        sns.scatterplot(data=line_df, x='image_number', y='inference_time', color=color, s=100, label=format)  # s=100 - размер точек
    else:
        print(f"Warning: No data found for format {format}. Skipping plot.")
        continue

    plt.xlabel('Image Number')
    plt.ylabel('Inference Time (ms)')
    plt.title(f'Inference Time vs. Image Number for {format}')
    plt.legend(title='Format')
    plt.grid(axis='y', linestyle='-', alpha=0.7)
    plt.grid(axis='x', linestyle=':', alpha=0.3)
    plt.ylim(3300, 7600)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Собираем данные в DataFrame
box_data = []
for library, lib_data in all_raw_results.items():
    for iteration, iter_data in lib_data.items():
        for result in iter_data[0]:
            if 'model_speed' in result and 'inference' in result['model_speed']:
                inference_time = result['model_speed']['inference']
                box_data.append({'library': library, 'inference_time': inference_time})
            else:
                print(f"Warning: Missing 'model_speed' or 'inference' in result for library {library}, iteration {iteration}")

box_df = pd.DataFrame(box_data)

plt.figure(figsize=(10, 6))
if not box_df.empty: # Добавляем проверку, чтобы избежать ошибки, если DataFrame пустой
    sns.boxplot(data=box_df, x='library', y='inference_time')
    plt.xlabel('Format')
    plt.ylabel('Inference Time (ms)')
    plt.title('Distribution of Inference Time (Box Plot)')

    # Добавляем ТОЛЬКО горизонтальную сетку:
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
else:
    print("Error: No data found for boxplot. Check the structure and content of 'all_raw_results'.")
```
